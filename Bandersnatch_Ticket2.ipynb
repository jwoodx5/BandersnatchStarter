```python
# Import necessary libraries
from pandas import DataFrame
import altair as alt

# Import the Database class from your data module
from app.data import Database

# Create an instance of the Database class with the collection name
db = Database('Database')

# Load data into a DataFrame
df = db.dataframe()

# View the first few rows of data
df.head()

# Get information about the DataFrame
df.info()

# Create a simple chart using Altair
chart = alt.Chart(df).mark_bar().encode(
    x='Your_Categorical_Column:N',
    y='count()'
).properties(
    title='Count of Items by Category'
).interactive()  # Add interactivity to the chart

# Display the chart in the notebook
chart
```
